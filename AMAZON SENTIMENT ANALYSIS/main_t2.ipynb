### Text processing

```python

def text_preprocessing(df, text_column):
    """Preprocesses text with the use of normalizing, cleaning, and also lemmatizing."""
    
    sw = set(stopwords.words('english'))  # Use a set for faster lookup
    word_freq = pd.Series(' '.join(df[text_column].dropna()).split()).value_counts()
    rare_words = set(word_freq[word_freq <= 1].index)

    def clean_text(text):
        text = str(text).lower()  # Lowercase conversion
        text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
        text = re.sub(r'\d+', '', text)  # Remove numbers
        text = ' '.join(word for word in text.split() if word not in sw)  # Remove stopwords
        text = ' '.join(word for word in text.split() if word not in rare_words)  # Remove rare words
        text = ' '.join(Word(word).lemmatize() for word in text.split())  # Lemmatization
        return text

    df[text_column] = df[text_column].astype(str).apply(clean_text)
    return df
```

```python
def text_visualization(df, text_column, barplot=True, wordcloud=True):
    """Visualizes using a bar plot and a word cloud."""
    
    # Compute Term Frequencies
    word_counts = pd.Series(" ".join(df[text_column].dropna()).split()).value_counts()

    if barplot:
        # Bar Plot of Most Frequent Words
        word_counts[word_counts > 1000].plot(kind="barh", figsize=(10, 6), color="skyblue", edgecolor="black")
        plt.title("Term Frequencies (Words appearing >1000 times)")
        plt.xlabel("Frequency")
        plt.ylabel("Words")
        plt.show()

    if wordcloud:
        # Generate Word Cloud
        text = " ".join(df[text_column].dropna())
        wc = WordCloud(max_font_size=100, max_words=1000, background_color="white").generate(text)

        plt.figure(figsize=(10, 6))
        plt.imshow(wc, interpolation="bilinear")
        plt.axis("off")
        plt.title("Word Cloud of Most Frequent Words")
        plt.show()

        wc.to_file("wordcloud.png")  # Save the word cloud image
```

### sentiment analysis

```python
def create_polarity_scores(df, text_column):
    """Adding polarity_score column using VADER sentiment analysis."""
    
    sia = SentimentIntensityAnalyzer()
    
    # Handle missing values before applying sentiment analysis
    df = df.dropna(subset=[text_column]).copy()
    
    # Compute polarity scores
    df["polarity_score"] = df[text_column].astype(str).apply(lambda x: sia.polarity_scores(x)["compound"])
    
    return df  # Return updated DataFrame
```

### Feature Engineering

```python
def create_label(df, text_column, label_column="sentiment_label"):
    """Creates sentiment labels ('pos' or 'neg') based on polarity scores and encodes them as 0 (neg) / 1 (pos)."""
    
    sia = SentimentIntensityAnalyzer()
    
    # Ensure text_column is string and handle missing values
    df = df.dropna(subset=[text_column]).copy()
    
    # Assign sentiment labels based on compound score
    df[label_column] = df[text_column].astype(str).apply(lambda x: "pos" if sia.polarity_scores(x)["compound"] > 0 else "neg")

    # Encode labels: 'pos' â†’ 1, 'neg' â†’ 0
    df[label_column] = LabelEncoder().fit_transform(df[label_column])

    return df[text_column], df[label_column] 
```

```python
def create_features_count(train_x, test_x):

    vectorizer = CountVectorizer()
    
    x_train_count_vectorizer = vectorizer.fit_transform(train_x)
    x_test_count_vectorizer = vectorizer.transform(test_x)  # Use transform, NOT fit_transform

    return x_train_count_vectorizer, x_test_count_vectorizer, vectorizer
```

```python
def create_features_TFIDF_word(train_x, test_x):

    tf_idf_word_vectorizer = TfidfVectorizer()
    
    x_train_tf_idf_word = tf_idf_word_vectorizer.fit_transform(train_x)
    x_test_tf_idf_word = tf_idf_word_vectorizer.transform(test_x) 

    return x_train_tf_idf_word, x_test_tf_idf_word, tf_idf_word_vectorizer
```

```python


def create_features_TFIDF_ngram(train_x, test_x):

    tf_idf_ngram_vectorizer = TfidfVectorizer(ngram_range=(2, 3))
    
    x_train_tf_idf_ngram = tf_idf_ngram_vectorizer.fit_transform(train_x)
    x_test_tf_idf_ngram = tf_idf_ngram_vectorizer.transform(test_x)  # Corrected!

    return x_train_tf_idf_ngram, x_test_tf_idf_ngram, tf_idf_ngram_vectorizer
```

```python


def create_features_TFIDF_chars(train_x, test_x):

    tf_idf_chars_vectorizer = TfidfVectorizer(analyzer="char", ngram_range=(2, 3))
    
    x_train_tf_idf_chars = tf_idf_chars_vectorizer.fit_transform(train_x)
    x_test_tf_idf_chars = tf_idf_chars_vectorizer.transform(test_x)  # Corrected!

    return x_train_tf_idf_chars, x_test_tf_idf_chars, tf_idf_chars_vectorizer
```

### Model creation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

def create_model_randomforest(train_x, test_x, train_y, test_y):

    models = {}
    accuracies = {}

    # Count Vectorizer
    x_train_count_vectorizer, x_test_count_vectorizer = create_features_count(train_x, test_x)
    rf_count = RandomForestClassifier()
    rf_model_count = rf_count.fit(x_train_count_vectorizer, train_y)
    accuracy_count = rf_model_count.score(x_test_count_vectorizer, test_y)
    print("Accuracy - Count Vectors: %.3f" % accuracy_count)
    
    models["count"] = rf_model_count
    accuracies["count"] = accuracy_count

    # TF-IDF Word
    x_train_tf_idf_word, x_test_tf_idf_word, _ = create_features_TFIDF_word(train_x, test_x)
    rf_word = RandomForestClassifier()
    rf_model_word = rf_word.fit(x_train_tf_idf_word, train_y)
    accuracy_word = rf_model_word.score(x_test_tf_idf_word, test_y)
    print("Accuracy - TF-IDF Word: %.3f" % accuracy_word)

    models["tfidf_word"] = rf_model_word
    accuracies["tfidf_word"] = accuracy_word

    # TF-IDF n-gram
    x_train_tf_idf_ngram, x_test_tf_idf_ngram, _ = create_features_TFIDF_ngram(train_x, test_x)
    rf_ngram = RandomForestClassifier()
    rf_model_ngram = rf_ngram.fit(x_train_tf_idf_ngram, train_y)
    accuracy_ngram = rf_model_ngram.score(x_test_tf_idf_ngram, test_y)
    print("Accuracy - TF-IDF n-gram: %.3f" % accuracy_ngram)

    models["tfidf_ngram"] = rf_model_ngram
    accuracies["tfidf_ngram"] = accuracy_ngram

    # TF-IDF Characters
    x_train_tf_idf_chars, x_test_tf_idf_chars, _ = create_features_TFIDF_chars(train_x, test_x)
    rf_chars = RandomForestClassifier()
    rf_model_chars = rf_chars.fit(x_train_tf_idf_chars, train_y)
    accuracy_chars = rf_model_chars.score(x_test_tf_idf_chars, test_y)
    print("Accuracy - TF-IDF Characters: %.3f" % accuracy_chars)

    models["tfidf_chars"] = rf_model_chars
    accuracies["tfidf_chars"] = accuracy_chars

    return models, accuracies
```
